A sample of graphs generated using Matplotlib and Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams.update({'font.size': 9})
```

```python
college_data = pd.read_csv('College_Data')
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(7,7))
plt.xlim(1000,9000)
plt.ylim(0,140)
sns.scatterplot(data=college_data,x='Room.Board',y='Grad.Rate',hue='Private',s=20,alpha=0.3)
leg = plt.legend()       #(markerscale=0.6,loc=(1,0.5),title='Private')
for lh in leg.legendHandles: lh.set_alpha(0.3)
```

Scatterplot of the costs of rooms rented by US students for accommodation and graduation rates for their respective universities. The colours of the dots divides the data into private (blue) and public (orange) universities

Below is the same graph, but with in a different grid

```python
sns.set_style('darkgrid')
plt.figure(figsize=(7,7))
plt.xlim(1000,9000)
plt.ylim(0,140)
sns.scatterplot(data=college_data,x='Room.Board',y='Grad.Rate',hue='Private',s=20,alpha=0.3)
leg = plt.legend()    # (markerscale=0.6,loc=(1,0.5),title='Private')
for lh in leg.legendHandles: lh.set_alpha(0.3)
```

```python
sns.set_style('darkgrid')
g = sns.FacetGrid(college_data, hue="Private", height=6, aspect=2, xlim=(0,25000))
g.map(plt.hist,"Outstate",histtype='barstacked',stacked=True,bins=20,alpha=0.3)
```

Histogram of US students that did their university tuitions outside their home states. The colours represent private (blue) and public (orange) universities

```python
iris = sns.load_dataset('iris')
```

```python
plt.figure(figsize=(7,5))
sns.kdeplot(data=iris[iris['species']=='setosa'],x='sepal_width',
            y='sepal_length',cmap='plasma',fill=True,thresh=0, levels=100)
```

A probability density map of sepal width against length for the *Iris setosa* plant

```python
from mpl_toolkits.mplot3d import axes3d
from matplotlib.ticker import LinearLocator

fig, ax = plt.subplots(subplot_kw={"projection": "3d"},figsize=(6.5,6.5),dpi=100)

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap='coolwarm', lw=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=3)
```
